```python
# Import goodies
import pandas as pd
import numpy as np
import math


# Import summoner names
data = pd.read_csv(r"C:\Users\toirt\LeagueOfLegendsPrediction\data\summoner_names.csv")
df = pd.DataFrame(data)
print(df.shape)
df.head()
```

```python
# League of Legends playerbase rank distribution as percents
rank_distribution = {
    'IRON IV': 0.03,
    'IRON III': 0.022,
    'IRON II': 0.011,
    'IRON I': 0.03,
    'BRONZE IV': 0.053,
    'BRONZE III': 0.047,
    'BRONZE II': 0.045,
    'BRONZE I': 0.036,
    'SILVER IV': 0.057,
    'SILVER III': 0.046,
    'SILVER II': 0.043,
    'SILVER I': 0.035,
    'GOLD IV': 0.064,
    'GOLD III': 0.048,
    'GOLD II': 0.048,
    'GOLD I': 0.04,
    'PLATINUM IV': 0.071,
    'PLATINUM III': 0.048,
    'PLATINUM II': 0.041,
    'PLATINUM I': 0.031,
    'EMERALD IV': 0.06,
    'EMERALD III': 0.029,
    'EMERALD II': 0.02,
    'EMERALD I': 0.012,
    'DIAMOND IV': 0.021,
    'DIAMOND III': 0.01,
    'DIAMOND II': 0.0064,
    'DIAMOND I': 0.0035,
    'MASTER I': 0.0042,
    'GRANDMASTER I': 0.0006,
    'CHALLENGER I': 0.00023
}
```

```python
# Need to filter this data to represent playerbase
def player_distribution(distribution, data):
    '''
    Takes in players and returns a list with players proportional to distribution of ranks
    '''
    total_players = (data.shape[0]*data.shape[1])/2 # split amount of players in half to account for percents greater than 3%
    names = []
    for key, value in distribution.items():
        if key in ['MASTER I', 'GRANDMASTER I', 'CHALLENGER I']: # Auto include 1 player from highest ranks
            names.append(df[key][:1])
        else:
            names.append(df[key][:round(value*total_players)]) # Get column from dataframe with rank and append desired players
    return names
                                          
# Apply function and convert to dataframe                     
names = player_distribution(rank_distribution, df)
names_df = pd.DataFrame(names).T

# Verifying distribution of names
names_df.head()

# Put all names from dataframe into one list
names_series = pd.concat([names_df[col] for col in names_df.columns])
names_list = names_series.dropna().tolist()
```

```python
# Export names as csv
df = pd.DataFrame(names_list, columns = ['SummonerName'])
df.to_csv(r"C:\Users\toirt\LeagueOfLegendsPrediction\data\final_summoner_names.csv", index = False)
```
